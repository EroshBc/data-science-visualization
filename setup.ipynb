```python
#examples on seaborne data sets
tips = sns.load_dataset("tips")
```

```python
tips
```

```python

ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust = .4)
plt.show()
```

```python
#flip the axis
sns.kdeplot(data=tips, y="total_bill")
plt.show()
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust = 5, cut=0)
plt.show()
```

```python
min(tips.total_bill)
```

```python
max(tips.total_bill)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue = "time")
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", hue = "time", multiple = "stack")
plt.show()
```

```python
#aggrevate values count, min, max,mean,sum,avg

tips_agg = (tips
            .groupby("size")
            .agg(total_bill = ("total_bill","mean"), n=("total_bill","count"))
                
            )

sns.kdeplot(data=tips_agg, x="total_bill", weights="n" )
plt.show()
```

```python
sns.kdeplot(
            data=tips, x="total_bill", hue ="size",
            fill=True, alpha=0.5, linewidth = 0)
plt.show()
```

```python
#rugplot- dataset on singular axis
import seaborn as sns
sns.set_theme()

tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", hue="time")
plt.show()
```
